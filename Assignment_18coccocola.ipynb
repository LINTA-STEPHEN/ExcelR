```python
Cola['Quarter_Year']=Cola['Quarter'].str.split('_').apply(lambda x: ' 19'.join(x[:]))
#Clean way to convert quarterly periods to datetime in pandas
Cola.head(3)
```

```python
# Converting into datetime formate as the index was not in correct formate.
Cola['Date']=(pd.to_datetime(Cola['Quarter_Year'].str.split(' ').apply(lambda x: ''.join(x[::-1])),dayfirst=True))
Cola.head(3)
```

```python
#Extracting the Quarters and putting in column
Cola['Quarters'] = 0

for i in range(42):
    n =Cola['Quarter'][i]
    Cola['Quarters'][i] = n[0:2]
    
Cola.head(3)
```

```python
#Creating Dummy Variables for Quarters Variable
dummy = pd.DataFrame(pd.get_dummies(Cola['Quarters']))
data= pd.concat((Cola, dummy), axis=1)

data.head()
```

```python
#Dropping all the unnecessary column
data.drop(['Quarter','Quarter_Year'] , axis =1, inplace = True)
```

```python
#Making the Date as DateTime index for the Dataframe
data.set_index('Date',inplace=True)
data.index.year
```

```python
#Separating Month and Year into Separate Column
Cola["Date"] = pd.to_datetime(Cola.Date,format="%b-%y")
Cola["month"] = Cola.Date.dt.strftime("%b") # month extraction
Cola["year"] = Cola.Date.dt.strftime("%Y") # year extraction
Cola.head(3)
```

**Sampling**

```python
#UpSampling with respect to month
upsampled_month=data.drop(['Quarters','Q1','Q2','Q3','Q4'], axis=1)
upsampled_month = upsampled_month.resample('M').mean()
print(upsampled_month.head())
```

```python
#fill missing values with the mean of the previous and next value
interpolated_month = upsampled_month.interpolate(method='linear')
print(interpolated_month.head())
interpolated_month.plot(figsize=(12,4))
plt.show()
```

```python
interpolated_month.head(3)
```

**Auto-Regressive Model**

```python
from statsmodels.tsa.ar_model import AutoReg
from statsmodels.tools.eval_measures import rmse
```

```python
# Split the data into train and test sets
train_size = int(len(interpolated_month) * 0.8)  # 80% for training
train, test = interpolated_month.iloc[:train_size], interpolated_month.iloc[train_size:]
```

```python
model = AutoReg(train['Sales'], lags=1)
model_fit = model.fit()
```

```python
# Make predictions
predictions = model_fit.predict(start=len(train), end=len(train)+len(test)-1, dynamic=False)
```

```python
# Evaluate the model
print("Root Mean Squared Error (RMSE):", rmse(test['Sales'], predictions))
```

```python
# Plot actual vs. predicted values
plt.figure(figsize=(10, 6))
plt.plot(train.index, train['Sales'], label='Train')
plt.plot(test.index, test['Sales'], label='Test')
plt.plot(test.index, predictions, label='Predictions')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('AR Model Forecast')
plt.legend()
plt.show()
```

```python
model_fit.summary()
```

**Moving Average**

```python
# Calculate moving averages
# Let's calculate a simple moving average over a window of 3 periods
window_size = 3
interpolated_month['moving_average'] = interpolated_month['Sales'].rolling(window=window_size).mean()

# Plot original data and moving averages
plt.figure(figsize=(10, 6))
plt.plot(interpolated_month.index, interpolated_month['Sales'], label='Original Data')
plt.plot(interpolated_month.index, interpolated_month['moving_average'], label='Moving Average (window={})'.format(window_size))
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Moving Averages')
plt.legend()
plt.show()
```

**ARMA & ARIMA**

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
# Fit ARMA model
arma_model = ARIMA(interpolated_month['Sales'], order=(2,0, 1))  # Example order (p, q)
arma_results = arma_model.fit()
```

```python
# Fit ARIMA model
arima_model = ARIMA(interpolated_month['Sales'], order=(1, 1, 1))  # Example order (p, d, q)
arima_results = arima_model.fit()
```

```python
# Make predictions
arma_predictions = arma_results.predict(start=0, end=len(interpolated_month)-1)
arima_predictions = arima_results.predict(start=0, end=len(interpolated_month)-1)
```

```python
# Plot original data and predictions
plt.figure(figsize=(10, 6))
plt.plot(interpolated_month.index, interpolated_month['Sales'], label='Original Data')
plt.plot(interpolated_month.index, arma_predictions, label='ARMA Predictions')
plt.plot(interpolated_month.index, arima_predictions, label='ARIMA Predictions')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('ARMA and ARIMA Model Predictions')
plt.legend()
plt.show()
```
